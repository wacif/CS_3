```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Input (size of the house in square feet)
X = np.array([1000, 1500, 2000, 2500, 3000]).reshape(-1, 1)

# Output (price of the house in dollars)
y = np.array([150000, 200000, 250000, 300000, 350000])
```

```python
# Create a linear regression model
model = LinearRegression()

# Train the model on the data
model.fit(X, y)
```

```python
# Predict the price of a 1200 sq. ft. house
predicted_price = model.predict([[1360]])
print(f"The predicted price for a 1200 sq. ft. house is: ${predicted_price[0]:,.2f}")
```

```python
# Plot the data points
plt.scatter(X, y, color='blue', label='Actual Data')

# Plot the regression line
plt.plot(X, model.predict(X), color='red', label='Regression Line')

# Add labels and legend
plt.xlabel('House Size (sq. ft.)')
plt.ylabel('House Price ($)')
plt.title('Linear Regression Example')
plt.legend()

plt.show()
```

# Implementation with Custom weights

```python
# Generate synthetic data (house sizes and prices)
np.random.seed(42)  # For reproducibility
X = np.random.randint(500, 5000, 100).reshape(-1, 1)  # House sizes (500 to 5000 sq. ft.)
y = (50 * X + 10000 + np.random.randint(-50000, 50000, X.shape)).flatten()  # Prices with noise

# Visualize the dataset
plt.scatter(X, y, color='blue', alpha=0.6, label='Generated Data')
plt.xlabel('House Size (sq. ft.)')
plt.ylabel('House Price ($)')
plt.title('Synthetic Dataset for Linear Regression')
plt.legend()
plt.show()
```

```python
# Train the Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Extract learned weights
m_learned = model.coef_[0]  # Slope (m)
b_learned = model.intercept_  # Intercept (b)
print(f"Learned Slope (m): {m_learned:.2f}")
print(f"Learned Intercept (b): {b_learned:.2f}")
```

```python
# Function to predict prices based on custom weights
def predict_custom(X, m, b):
    return m * X + b

# Interactive weights for demonstration
custom_m = float(input("Enter custom slope (m): "))
custom_b = float(input("Enter custom intercept (b): "))

# Make predictions using custom weights
y_custom = predict_custom(X, custom_m, custom_b)

# Visualize custom line vs learned line
plt.scatter(X, y, color='blue', alpha=0.6, label='Generated Data')
plt.plot(X, model.predict(X), color='red', label='Learned Line')
plt.plot(X, y_custom, color='green', linestyle='--', label='Custom Line')
plt.xlabel('House Size (sq. ft.)')
plt.ylabel('House Price ($)')
plt.title('Comparison of Learned Line vs Custom Line')
plt.legend()
plt.show()
```

# Prediction Comparison with Custom weights

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate synthetic data
np.random.seed(42)
X = np.random.randint(500, 5000, 100).reshape(-1, 1)  # House sizes (sq. ft.)
y = (50 * X + 10000 + np.random.randint(-50000, 50000, X.shape)).flatten()  # Prices with noise

# Train the Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Extract learned weights
m_learned = model.coef_[0]
b_learned = model.intercept_
print(f"Learned Slope (m): {m_learned:.2f}")
print(f"Learned Intercept (b): {b_learned:.2f}")
```

```python
# Predict with learned weights
def predict_learned(size):
    return m_learned * size + b_learned

# Predict with custom weights
def predict_custom(size, custom_m, custom_b):
    return custom_m * size + custom_b
```

```python
# Input custom weights
custom_m = float(input("Enter custom slope (m): "))
custom_b = float(input("Enter custom intercept (b): "))
```

```python
# Example house size for prediction
house_size = float(input("Enter the size of the house (sq. ft.) to predict price: "))

# Predictions
price_learned = predict_learned(house_size)
price_custom = predict_custom(house_size, custom_m, custom_b)

# Print results
print(f"Predicted price using the learned line: ${price_learned:,.2f}")
print(f"Predicted price using the custom line: ${price_custom:,.2f}")
```

```python
# Plot the dataset
plt.scatter(X, y, color='blue', alpha=0.6, label='Data Points')

# Plot the learned line
plt.plot(X, model.predict(X), color='red', label='Learned Line')

# Plot the custom line
y_custom = predict_custom(X, custom_m, custom_b)
plt.plot(X, y_custom, color='green', linestyle='--', label='Custom Line')

# Highlight prediction points
plt.scatter(house_size, price_learned, color='red', label='Prediction (Learned)', zorder=5)
plt.scatter(house_size, price_custom, color='green', label='Prediction (Custom)', zorder=5)

# Add labels and legend
plt.xlabel('House Size (sq. ft.)')
plt.ylabel('House Price ($)')
plt.title('Comparison of Learned and Custom Predictions')
plt.legend()
plt.show()
```
